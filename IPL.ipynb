## IPL Dataset Kaggle

```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
match_filepath = "IPL_Matches_2008-2020.csv"

match_data = pd.read_csv(match_filepath)
```

```python
match_data.head()
```

```python
match_df = pd.DataFrame(match_data)
```

```python
x=['Royal Challengers Bangalore', 'Kolkata Knight Riders', 'Kings XI Punjab', 'Rajasthan Royals', 'Mumbai Indians', 
   'Chennai Super Kings', 'Sunrisers Hyderabad', 'Delhi Capitals', 'Delhi Daredevils', 'Rising Pune Supergiant',
   'Gujarat Lions', 'Pune Warriors', 'Kochi Tuskers Kerala', 'Deccan Chargers', 'Rising Pune Supergiants']

y = ['RCB','KKR','KXIP','RR','MI','CSK','SRH','DC','DC','RPS','GL','PWI','KTK','DCH', 'RPS']
```

```python
match_df.replace(x,y,inplace = True)

match_df
```

```python
match_df.info()
```

```python
match_df.isnull().sum()
```

### Total no of matches in one season

```python
match_season = match_df['date'].str[:4].astype(int)
match_season
```

```python
match_season = match_season.value_counts().reset_index()
match_season
```

```python
plt.figure(figsize=(10,6))

plt.title("Matches Every Season")

sns.barplot(x=match_season['index'], y=match_season['date'])

plt.ylabel("No of Matches")
plt.xlabel("Season")
```

### Matches won by teams

```python
match_win = match_df['winner'].value_counts().reset_index()
match_win
```

```python
plt.figure(figsize = (10,6))

sns.barplot(x = match_win['index'], y= match_win['winner'], palette='rocket')

plt.ylabel('No of Wins')
plt.xlabel('Team')
plt.title('Matches won by the Teams')
```

### Stadiums with most no of matches (top 10)

```python
plt.figure(figsize = (12,8))

sns.countplot(y = 'venue',
              data = match_df,
              order = match_df['venue'].value_counts().iloc[:10].index,
              palette='deep')

plt.xlabel('No: of matches')
plt.ylabel('Venue')
plt.title('Number of matches played in each Stadium')
```

### Win percentage

```python
matches_played=pd.concat([match_data['team1'],match_data['team2']])
matches_played=matches_played.value_counts().reset_index()
```

```python
matches_played
```

```python
matches_played=pd.concat([match_data['team1'],match_data['team2']])
matches_played=matches_played.value_counts().reset_index()
matches_played.columns=['Team','Total Matches']
matches_played['wins']=match_data['winner'].value_counts().reset_index()['winner']

matches_played.set_index('Team',inplace=True)
totm = matches_played.reset_index().head(8)
```

```python
matches_played
```

```python
matches_played.index
```

```python
matches_played['win percent'] = matches_played['wins']/matches_played['Total Matches']*100
```

```python
matches_played
```

```python
plt.figure(figsize=(10,6))

plt.title("Win percentage of each team")

sns.barplot(x = matches_played.index, y = matches_played['win percent'], palette='deep')
```

### Win percentage of toss winning team

```python
matches_toss = pd.concat([match_data['toss_winner']])
matches_toss = matches_toss.value_counts().reset_index()
```

```python
home = match_df.loc[match_df['toss_winner'] == match_df['winner']]
home.reset_index()
```

```python
home['winner']
```

```python
matches_toss.columns=['Team','No of Toss Wins']
matches_toss['wins']=home['winner'].value_counts().reset_index()['winner']

matches_toss.set_index('Team',inplace=True)
```

```python
matches_toss
```

```python
matches_toss['win percent'] = matches_toss['wins']/matches_toss['No of Toss Wins']*100
matches_toss
```

```python
plt.figure(figsize=(10,6))

plt.title("Win percentage of toss winning team")

clrs = ['orange']
sns.barplot(x = matches_toss.index, y = matches_toss['win percent'], palette=clrs)
```
